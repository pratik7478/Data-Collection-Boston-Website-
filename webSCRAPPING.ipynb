# WRAPING THE INFORMATION OF THE jobs

# " p " tag 


```python
jobs=soup.find_all("p",{"class":"result-info"})
```

```python
jobs
```

```python
#we want to extract the each detial of the job so 
#AND FOR LOACTION ,LOCATIO IS ASIGN FOR ONLY FEW SO IF TAG MUST BE THERE ((IF YES ok / IF NO THEN NA))
for job in jobs:
    title=job.find('a',{'class':"result-title"}).text
    location_tag=job.find('span',{'class':"result-hood"})
    location=location_tag.text[2:-1] if location_tag else "N/A"
    date=job.find("time",{'class':"result-date"}).text
    link=job.find("a",{'class':"result-title"}).get("href")

    print("JOB_TITLE:",title,"\nLOCATION:",location,"\nDATE:",date,"\nLINK:",link,"\n---")
```

#  open each page of the job and extract the information

```python
#we want to extract the each detial of the job so 
#AND FOR LOACTION ,LOCATIO IS ASIGN FOR ONLY FEW SO IF TAG MUST BE THERE ((IF YES ok / IF NO THEN NA))
for job in jobs:
    title=job.find('a',{'class':"result-title"}).text
    location_tag=job.find('span',{'class':"result-hood"})
    location=location_tag.text[2:-1] if location_tag else "N/A"
    date=job.find("time",{'class':"result-date"}).text
    link=job.find("a",{'class':"result-title"}).get("href")
    
    job_response=requests.get(link)
    job_data=job_response.text
    job_soup=BeautifulSoup(job_data,'html.parser')
    job_desc=job_soup.find("section",{"id":"postingbody"}).text
    job_att_tag=job_soup.find('p',{"class":"attrgroup"})
    job_attribute=job_att_tag.text if job_att_tag else "N/A"

    print("JOB_TITLE:",title,"\nLOCATION:",location,"\nDATE:",date,"\nLINK:",link,"\nJOB_DESC",job_desc,"\nJOB-ATTRIBUTE",job_attribute,"\n---")
```

# moving to the next page of same webpage

# CRAWLING AND SCRAPING NEXT PAGE



```python
job_no=0
while True:
    response=requests.get(url)
    data=response.text
    soup=BeautifulSoup(data,"html.parser")
    jobs=soup.find_all('p',{"class":"result-info"})

    for job in jobs:
        title=job.find('a',{'class':"result-title"}).text
        location_tag=job.find('span',{'class':"result-hood"})
        location=location_tag.text[2:-1] if location_tag else "N/A"
        date=job.find("time",{'class':"result-date"}).text
        link=job.find("a",{'class':"result-title"}).get("href")
    
        job_response=requests.get(link)
        job_data=job_response.text
        job_soup=BeautifulSoup(job_data,'html.parser')
        job_desc=job_soup.find("section",{"id":"postingbody"}).text
        job_att_tag=job_soup.find('p',{"class":"attrgroup"})
        job_attribute=job_att_tag.text if job_att_tag else "N/A"
        job_no+=1
        print("JOB_TITLE:",title,"\nLOCATION:",location,"\nDATE:",date,"\nLINK:",link,"\nJOB_DESC",job_desc,"\nJOB-ATTRIBUTE",job_attribute,"\n---")

    url_tag=soup.find('a',{'title':'next page'})
    if url_tag.get('href'):
        url=""+url_tag.get('href')
        print(url)
    else:
        break
        
print("NO OF JOBS=",job_no)            
```

# SAVE THE OUTPUT TO SAVE THE FILE IN CSV

```python
import pandas as pd
```

```python
#create a empty dictonary to save the data

npo_job={}
job_no=0
while True:
    response=requests.get(url)
    data=response.text
    soup=BeautifulSoup(data,"html.parser")
    jobs=soup.find_all('p',{"class":"result-info"})

    for job in jobs:
        title=job.find('a',{'class':"result-title"}).text
        location_tag=job.find('span',{'class':"result-hood"})
        location=location_tag.text[2:-1] if location_tag else "N/A"
        date=job.find("time",{'class':"result-date"}).text
        link=job.find("a",{'class':"result-title"}).get("href")
    
        job_response=requests.get(link)
        job_data=job_response.text
        job_soup=BeautifulSoup(job_data,'html.parser')
        job_desc=job_soup.find("section",{"id":"postingbody"}).text
        job_att_tag=job_soup.find('p',{"class":"attrgroup"})
        job_attribute=job_att_tag.text if job_att_tag else "N/A"
        job_no+=1
        #updating the dictonary
        npo_job[job_no]=[title,location,date,link,job_desc,job_attribute]
        print("JOB_TITLE:",title,"\nLOCATION:",location,"\nDATE:",date,"\nLINK:",link,"\nJOB_DESC",job_desc,"\nJOB-ATTRIBUTE",job_attribute,"\n---")

    url_tag=soup.find('a',{'title':'next page'})
    if url_tag.get('href'):
        url=""+url_tag.get('href')
        print(url)
    else:
        break
        
print("NO OF JOBS=",job_no)            


npo_job_df=pd.DataFrame.from_dict(npo_job,orient="index",columns=['JOB TITLE','LOCATION','DATE','LINK',"JOB_DESCRIPTION","JOB_ATTRIBUTE"])
```
